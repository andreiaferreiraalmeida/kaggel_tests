```python
def upsample(df_aux, df_col):
    for col in df_aux.columns:
        if df_col in col:
            df_aux_count = df_aux[col].value_counts()
            df_aux_count_max = df_aux_count.max()
            df_aux_sample_final = []

            for i, v in df_aux_count.items():
                df_aux_sample = df_aux[df_aux[col] == i]
                df_aux_sample = resample(df_aux_sample, replace=True, n_samples=df_aux_count_max, random_state=27)
                df_aux_sample_final.extend([df_aux_sample])

            return pd.concat(df_aux_sample_final)
```

### RSM - Multinomial naive Bayes

```python
def custom_analyzer(data):
    result = re.sub('[?]+', ' ', data).split()
    
    return result
```

```python
v = TfidfVectorizer(analyzer=custom_analyzer,lowercase=True)
print(v.fit(['N?VCI ??ICIMC?']).vocabulary_)
```

```python
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(analyzer=custom_analyzer, lowercase=True)),
    ('classifier', MultinomialNB())
])
```

```python
# Text Splitter using ? as delimiter
def text_splitter(data):
    if isinstance(data, float):
        return ()

    return re.sub('[?]+', ' ', data).split()


# Text Length without spaces
def text_length(data):
    return np.array([len(v.replace(' ', '')) for v in data]).reshape(-1, 1)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import FeatureUnion
from sklearn.preprocessing import FunctionTransformer

class SubjectBodyExtractor(BaseEstimator, TransformerMixin):
    def fit(self, x, y=None):
        return self

    def transform(self, df):
        features = np.recarray(shape=len(df.index), dtype=[('p11_mot', object)])
        i = 0
        for _, row in df.iterrows():
            features['p11_mot'][i] = row['IND_MOT_P11_07_RL']
            i += 1

        return features

# Returns items from specific dictionary
class ItemSelector(BaseEstimator, TransformerMixin):
    def __init__(self, key):
        self.key = key

    def fit(self, x, y=None):
        return self

    def transform(self, data_dict):
        return data_dict[self.key]

# Multinomial NB declaration
pipeline = Pipeline([
    # Extracts the subject & body
    ('subjectbody', SubjectBodyExtractor()),

    # Uses FeatureUnion to combine features
    ('union', FeatureUnion(
        transformer_list=[
            # Pipeline for P11 Description
            ('p11_text', Pipeline([
                ('selector', ItemSelector(key='p11_mot')),
                ('tfidf', TfidfVectorizer(analyzer=text_splitter, lowercase=True)),
            ])),
            # Pipeline for P11 Length
            ('p11_len', Pipeline([
                ('selector', ItemSelector(key='p11_mot')),
                ('func', FunctionTransformer(text_length, validate=False))
            ]))
        ],

        # Defines weight for components in FeatureUnion
        transformer_weights={
            'p11_text': 1.0,
            'p11_len': 1.0,
        },
    )),

    # Use the classifier on the combined features
    ('classifier', MultinomialNB())
])
```

```python
from difflib import SequenceMatcher

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()

similar("Apple","Appel")
```

```python
def lookup(row):
    best_value = 0
    best_desc = ''
    for _, rel in religioes_excel[religioes_excel['IND_COD_P11_07_RL'] == row['IND_NLP_P11_07_RL']].dropna().iterrows():
        value = similar(rel['IND_MOT_P11_07_RL'], row['IND_MOT_P11_07_RL'])
        if value > best_value:
            best_value = value
            best_desc = rel['IND_MOT_P11_07_RL']
    return best_desc
# Creates a new column for Description
results.reset_index(drop=True)
results['RELIGIOES_TXT'] = results.apply(lambda row: lookup(row), axis=1)
```
